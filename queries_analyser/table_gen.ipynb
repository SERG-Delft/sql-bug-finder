```python
rule = 0
cnt = 0
print(header, end='')
for index, row in df.iterrows():
    num = row[0]
    rule_id = row[2]
    query = row[3]
    query = query.replace('%', '\%')
    query = query.replace('$', '\$')
    query = query.replace('#', '\#')
    query = (query[:300] + '...') if len(query) > 300 else query
    correct = '\\emph{YES}' if row[5] is True else '\\emph{NO}'
    query_col = '\\multicolumn{2}{X}{\sql{' + query + '}}'
    print(f'{num} & {rule_id} & {query_col} & {correct} \\\\ \\hline')
    cnt += 1
    if cnt == 20:
        cnt = 0
        caption = '\\caption{Manual analysis of tool performance for rule ' + rules[rule] + '}'
        print(caption, end='')
        print(footer)
        print(header, end='')
        rule += 1
```
